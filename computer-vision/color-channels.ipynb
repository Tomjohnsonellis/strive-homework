### To edit an image making a copy of it is always a good idea!

```python
plt.figure(figsize=(10,10))
# Load an image
original = cv2.imread("color-test.png")
# And make a copy of the original image to avoid manipulating the original image
image = original.copy()
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

## Color Spaces

### BGR Spliting the color channels

```python
original = cv2.imread("color-test.png", cv2.IMREAD_COLOR)
image = original.copy()
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Load a color image and split in is different B, G and R channels
image.shape
R,G,B = cv2.split(image)
plt.imshow(R)
```

[Hint](# "cv2.split") 

```python
# Show each channel channel on screen
cv2.imshow("Red", R)
cv2.waitKey(0)
cv2.imshow("Green", G)
cv2.waitKey(0)
cv2.imshow("Blue", B)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Merging the color channels

```python
# Modify a channel and then merge them all again
# I'll remove blue


R,G,B = cv2.split(image)
B = B+100
extra_blue = cv2.merge([R,G,B])
plt.imshow(extra_blue)
```

[Hint](# "cv2.merge") 

```python
# Create a black canvas with the same size as the image you loaded
print(f"Image shape: {image.shape}")
black = np.full_like(image, 0)
plt.imshow(black)
print(f"Black shape: {black.shape}")
black = np.zeros(image.shape[:2], dtype="uint8")
```

[Hint](# "use numpy np.zeros function with the img.shape[:2]") 

```python
# Merge each channel with the black canvas 
# So you will end up with 3 images: (Red,Black,Black), (Black,Green,Black) and (Black,Black,Blue) 
# Assuming you are working with RGB, but BGR is also fine

black_red = cv2.merge([R, black, black])
black_green = cv2.merge([black, G, black])
black_blue = cv2.merge([black, black, B])
plt.imshow(black_red)
```

[Hint](# "cv2.merge")

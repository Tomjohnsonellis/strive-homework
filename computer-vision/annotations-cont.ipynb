# Computer Vision Image annotations

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cv2.__version__
```

### To edit an image making a copy of it is always a good idea!

```python
# Load an imaged
original = cv2.imread("test.jpg")
# And make a copy of the original image to avoid manipulating the original image
image = original.copy()
# Convert to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

[Hint](# "Numpy .copy function") 

## Drawing

### Drawing lines

```python
# Draw some lines on top of the image

cv2.line(image, (100,200), (100,800), color=(255,0,255), thickness=50)
cv2.line(image, (100,500), (200,500), color=(0,255,0), thickness=10)
cv2.line(image, (200,500), (400,500), color=(0,255,225), thickness=25)
plt.imshow(image)
```

[Hint](# "cv2.line function") 

### Circles

```python
# Draw some circles on top of the image
cv2.circle(image, (250,600), radius=10, color=(255,0,0), thickness=(10))
cv2.circle(image, (500,500), radius=1, color=(0,0,255), thickness=50)
cv2.circle(image, (500,500), radius=40, color=(0,0,128), thickness=10)
plt.imshow(image)
```

[Hint](# "cv2.circle") 

### Ellipses

```python
# Draw some elipses on top of the image
cv2.ellipse(image, (200,800), (25,150), angle=90, startAngle=0, endAngle=360, color=(128,128,0), thickness=100)
plt.imshow(image)
```

[Hint](# "cv2.ellipse") 

### Poligons

```python
# Lets draw some poligons, first declare the points of the poligon
points = np.array( [[800,600], [800,900], [600,900], [600,600]])
points = points.reshape((-1,1,2))
```

[Hint](# "Example: np.array([[10,5],[120,130],[170,120],[250,210]], np.int32)") 

```python
# Draw the poligon on top of the image
cv2.polylines(image, [points], True, color=(128,0,128),thickness=2 )
plt.imshow(image)
```

[Hint](# "cv2.polylines(img_copy,[pts],False,(255,0,0),3)")  

### Add an overlay on top of an image

```python
# Load an image and create a copy of it you can call it overlay
original = cv2.imread("cropped_apples.jpg")
apples = original.copy()
overlay = apples.copy()
# Create a rectangle with a fill color (-1) on top of the overlay image
cv2.rectangle(overlay, (200,10), (390, 260), color=(0,128,120), thickness=-1)

# Combine both images setting some opacity on the overlay
stack = cv2.addWeighted(overlay, alpha=0.5, src2=apples, beta=0.5, gamma=0)
# Better method
alpha = 0.75
stack = cv2.addWeighted(overlay, alpha=alpha, src2=apples, beta=1-alpha, gamma=0)
plt.imshow(stack)
# Add some text on top of the overlay and show the image on screen
```
